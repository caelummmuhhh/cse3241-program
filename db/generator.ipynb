```python
def generate_random_phone_number():
    area_code = random.randint(100, 999)  # Area code (3 digits)
    central_office_code = random.randint(100, 999)  # Central office code (3 digits)
    line_number = random.randint(1000, 9999)  # Line number (4 digits)
    
    phone_number = f'+1 ({area_code}) {central_office_code}-{line_number}'
    return phone_number

def generate_random_email(fname: str, lname: str):
    randInt = random.randint(0, 4)
    domains = [
        'yahoo.com',
        'gmail.com',
        'hotmail.com',
        'osu.edu',
        'buckeyemail.osu.edu',
        'ymail.com',
        'aol.com',
        'cox.net',
        'verizon.net',
        'terra.com.br',
        'hotmail.es',
        'icloud.com',
        'yahoo.co.uk',
    ]
    if randInt == 2:
        return f'{emails[random.randint(0, len(emails)-1)].split('@')[0]}@{domains[random.randint(0, len(domains)-1)]}'
    if randInt == 0:
        return f'{fname.lower()}.{lname.lower()}.{random.randint(0, 9999)}@{domains[random.randint(0, len(domains)-1)]}'
    if randInt == 1:
        return f'{fname.lower()}_{lname.lower()}{random.randint(0, 9999)}@{domains[random.randint(0, len(domains)-1)]}'
    if randInt == 3:
        return f'{fname.capitalize()}{lname.capitalize()}{random.randint(0, 9999)}@{domains[random.randint(0, len(domains)-1)]}'
    return f'{lname.capitalize()}{random.randint(0, 999999)}@{domains[random.randint(0, len(domains)-1)]}'

def generate_random_date(start: date, end: date):
    diff = end - start
    randDays = random.randint(0, diff.days)
    return start + timedelta(days=randDays)
```
